2000년, 2005년, 2010년, 2015년, 2020년 대한민국의 남녀 인구가 각각 남 [23.5, 23.9, 24.3, 24.6, 25.2], 여 [22.7, 23.1, 23.6, 24.0, 24.8] 이다.

해당 데이터를 그래프로 시각화하려 한다. 

그래프의 제목은 Population (millions) 이고 선의 두깨는 5, 선의 모양은 대쉬, 마커는 점, 마커의 크기는 10으로 지정하며

남자 그래프는 선의 색상은 #0000ff50, 마커의 색상은 #0000ff으로 지정하고

여자 그래프는 선의 색상은 #ff000050, 마커의 색상은 #ff0000으로 지정하하여 아래와 같이 표현하려고 한다.

이때 이 python의 matplotlib의 pyplot을 이용하여 그래프를 그리는 코드를 작성하시오.

```python
import matplotlib.pyplot as plt

%matplotlib inline
year = [2000, 2005, 2010, 2015, 2020]
Male = [23.5, 23.9, 24.3, 24.6, 25.2]
Female = [22.7, 23.1, 23.6, 24.0, 24.8]
plt.title('Population (millions)')
plt.plot(year, Male, c='#0000ff50', mec='#0000ff', lw=5, ls=':', marker='.', ms=10, mfc='#0000ff' ,label = 'Male')
plt.plot(year, Female, c='#ff000050', mec='#ff0000',lw=5, ls=':', marker='.', ms=10,mfc='#ff0000' ,label = 'Female')
plt.grid(True)
plt.legend(loc=4)
plt.show()
```
